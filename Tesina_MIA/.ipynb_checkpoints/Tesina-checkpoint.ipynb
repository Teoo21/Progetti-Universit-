```python
import pandas as pd
import os
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from FisherDA import MultipleFisherDiscriminantAnalysis as MDA
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
columns = [
    "compactness", "circularity", "distance_circularity", "radius_ratio", 
    "pr_axis_aspect_ratio", "max_length_aspect_ratio", "scatter_ratio", 
    "elongatedness", "pr_axis_rectangularity", "max_length_rectangularity", 
    "scaled_variance_major", "scaled_variance_minor", "scaled_radius_of_gyration", 
    "skewness_major", "skewness_minor", "kurtosis_minor", "kurtosis_major", 
    "hollows_ratio", "vehicle_class"
]

file_list = ['xaa.dat', 'xab.dat', 'xac.dat', 'xad.dat', 'xae.dat', 'xaf.dat', 'xag.dat', 'xah.dat', 'xai.dat']

dataframes = []
for file in file_list: #Per fare il ciclo for mi sono aiutato con AI per capire come manipolare correttamente i file .dat
    if os.path.exists(file):
        df_temp = pd.read_csv(file, sep='\s+', names=columns, header=None)
        dataframes.append(df_temp)

df_veicoli = pd.concat(dataframes, ignore_index=True)
df_veicoli_copy = df_veicoli.copy()
display(df_veicoli.head())
```

```python
df_veicoli.info()
```

```python
le = LabelEncoder()
df_veicoli_copy['vehicle_class'] = le.fit_transform(df_veicoli_copy['vehicle_class'])
df_veicoli_copy.info()
```

```python
df_veicoli_copy.describe()
```

```python
df_bilancio = df_veicoli['vehicle_class'].value_counts().reset_index()
df_bilancio.columns = ['Classe', 'Numerosità']

totale = df_bilancio['Numerosità'].sum()
df_bilancio['Frequenza'] = (df_bilancio['Numerosità'] / totale * 100).round(2)

df_bilancio
```

```python
X = df_veicoli.drop('vehicle_class',axis=1)
y = df_veicoli_copy['vehicle_class']

# Per prima cosa suddivido il dataset in training set e test set per non avere problemi di bias successivamente
# Uso stratify in modo tale da mantenere la percentuale di ogni classe nei vari set bilanciata

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42, stratify=y)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# PCA sui dati non scalati
pca_vehicle_nostd = PCA().fit(X_train)

#PCA sui dati scalati
pca_vehicle_std_train = PCA().fit(X_train_scaled)

# La scelta di fare la PCA sul training set è solo per poi poter visualizzare i dati di training con cui addestriamo i metodi

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 14))

def scree(ax, pca, title):
    n_feats = pca.n_features_in_
    ratio = pca.explained_variance_ratio_
    cum_ratio = np.cumsum(ratio)
    
    ax.bar(np.arange(n_feats), ratio, bottom=np.insert(cum_ratio, 0, 0)[:-1])
    ax.bar(np.arange(n_feats), np.insert(cum_ratio, 0, 0)[:-1], color='b', alpha=0.15)
    ax.plot(cum_ratio, 'r', marker='o')
    
    ax.set_title(title, fontsize=14)
    ax.set_xticks(np.arange(n_feats))
    ax.set_xticklabels([f'PC{i}' for i in range(1, n_feats + 1)])
    ax.set_xlabel('Principal components')
    ax.set_ylabel('Varianza spiegata cumulata (%)')
    ax.grid(True, linestyle='--', alpha=0.6)
    ax.set_ylim([0, 1.05])

scree(ax1, pca_vehicle_nostd, 'VEHICLE (SENZA STANDARDIZAZIONE)')
scree(ax2, pca_vehicle_std_train, 'VEHICLE (CON STANDARDIZAZIONE)')

plt.tight_layout(pad=3.0)
plt.show()
```

```python
feature_labels = X_train.columns.to_list()
n_features = pca_vehicle_std_train.n_features_in_

plt.figure(figsize=(10, 6))
plt.bar(np.arange(n_features), pca_vehicle_std_train.components_[0, :])
plt.xticks(ticks=np.arange(n_features), labels=feature_labels, rotation=45, ha='right')
plt.title('VEHICLE - PC1')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.bar(np.arange(n_features), pca_vehicle_std_train.components_[1, :])
plt.xticks(ticks=np.arange(n_features), labels=feature_labels, rotation=45, ha='right')
plt.title('VEHICLE - PC2')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.bar(np.arange(n_features), pca_vehicle_std_train.components_[2, :])
plt.xticks(ticks=np.arange(n_features), labels=feature_labels, rotation=45, ha='right')
plt.title('VEHICLE - PC3')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
X_pca_train = pca_vehicle_std_train.transform(X_train_scaled)

pc1 = X_pca_train[:, 0]
pc2 = X_pca_train[:, 1]
pc3 = X_pca_train[:, 2]

target_names = {0: 'bus', 1: 'opel', 2: 'saab', 3: 'van'}

plt.figure(figsize=(10, 7))
for c in np.unique(y_train):
    mask = (y_train == c)
    plt.scatter(X_pca_train[mask, 0], X_pca_train[mask, 1], label=target_names[c], alpha=0.6, s=50)

plt.title('PCA 2D - Prime 2 Componenti')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

for c in np.unique(y_train):
    mask = (y_train == c)
    ax.scatter(X_pca_train[mask, 0], X_pca_train[mask, 1], X_pca_train[mask, 2], label=target_names[c], alpha=0.6, s=50)

ax.set_title('PCA 3D - Prime 3 Componenti')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.legend()

ax.view_init(elev=20, azim=45)

plt.show()
```

```python
#Scelgo di tenermi da parte i dati proiettati in 10 componenti principali, così da avere più del 98% di varianza spiegata mantenuta

pca_vehicle_std_train = PCA(n_components = 10).fit(X_train_scaled)
X_pca_train = pca_vehicle_std_train.transform(X_train_scaled)

X_pca_test = pca_vehicle_std_train.transform(X_test_scaled)

#questi verrano utilizzati più avanti
```

```python
#Ho scleto di fittare il metodo MDA sui dati di train, corenentemente con quanto fatto per la PCA, così da poter visualizzare i dati di training con cui addestro i metodi

mda = MDA(n_dimensions=3)

mda.fit(X_train_scaled, y_train)
Zmda_train = mda.transform(X_train_scaled).real #Inizialmente senza metodo .real mi dava errore, cercando ho capito che i dati potevano avere una parte immaginaria infinitesima dovuta a errori di computazione, che in questo modo non considero
Zmda_test = mda.transform(X_test_scaled).real #Mi serviranno più avanti

fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

target_names = {0: 'bus', 1: 'opel', 2: 'saab', 3: 'van'}

for c in np.unique(y_train):
    mask = (y_train == c)
    ax.scatter(Zmda_train[mask, 0], Zmda_train[mask, 1], Zmda_train[mask, 2], label=target_names[c], alpha=0.6, s=50)

ax.set_title('MDA in 3D')
ax.legend(title="Veicoli")

ax.view_init(elev=20, azim=45)

plt.show()
```

```python
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(111)

target_names = {0: 'bus', 1: 'opel', 2: 'saab', 3: 'van'}

for c in np.unique(y_train):
    mask = (y_train == c)
    ax.scatter(Zmda_train[mask, 0], Zmda_train[mask, 1], label=target_names[c], alpha=0.6, s=50)

ax.set_title('MDA in 2D')
ax.legend(title="Veicoli")
```

```python
k_neighbors = range(1, 51)

acc_train_raw = []
acc_test_raw  = []
acc_test_pca  = []
acc_test_mda  = []

for k in k_neighbors:
    # Predizione sui dati raw
    knn_raw = KNN(n_neighbors=k)
    knn_raw.fit(X_train_scaled, y_train)
    acc_train_raw.append(accuracy_score(y_train, knn_raw.predict(X_train_scaled)))
    acc_test_raw.append(accuracy_score(y_test, knn_raw.predict(X_test_scaled)))
    
    #Predizione sui dati con PCA
    knn_pca = KNN(n_neighbors=k)
    knn_pca.fit(X_pca_train, y_train)
    acc_test_pca.append(accuracy_score(y_test, knn_pca.predict(X_pca_test)))
    
    #Predizioni sui dati con MDA
    knn_mda = KNN(n_neighbors=k)
    knn_mda.fit(Zmda_train, y_train)
    acc_test_mda.append(accuracy_score(y_test, knn_mda.predict(Zmda_test)))

plt.figure(figsize=(15, 6))

plt.plot(k_neighbors, acc_train_raw, label='KNN sul Train set')
plt.plot(k_neighbors, acc_test_raw,  label='KNN sul Test set')
plt.plot(k_neighbors, acc_test_pca,  label='KNN sul Test set (Con PCA)')
plt.plot(k_neighbors, acc_test_mda,  label='KNN sul Test set (Con MDA)')

plt.title('Analisi dell\'Accuracy del KNN al variare di k (voti non pesati)')
plt.xlabel('Numero di vicini (k)')
plt.ylabel('Accuracy')
plt.xticks(np.arange(0, 51, 5))
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='upper right')

plt.show()
```

```python
k_neighbors = range(1, 51)

acc_train_raw = []
acc_test_raw  = []
acc_test_pca  = []
acc_test_mda  = []

for k in k_neighbors:
    # Predizione sui dati raw
    knn_raw = KNN(n_neighbors=k, weights='distance')
    knn_raw.fit(X_train_scaled, y_train)
    acc_train_raw.append(accuracy_score(y_train, knn_raw.predict(X_train_scaled)))
    acc_test_raw.append(accuracy_score(y_test, knn_raw.predict(X_test_scaled)))
    
    #Predizione sui dati con PCA
    knn_pca = KNN(n_neighbors=k, weights='distance')
    knn_pca.fit(X_pca_train, y_train) 
    acc_test_pca.append(accuracy_score(y_test, knn_pca.predict(X_pca_test)))
    
    #Predizioni sui dati con MDA
    knn_mda = KNN(n_neighbors=k, weights='distance')
    knn_mda.fit(Zmda_train, y_train)
    acc_test_mda.append(accuracy_score(y_test, knn_mda.predict(Zmda_test)))

plt.figure(figsize=(15, 6))

plt.plot(k_neighbors, acc_train_raw, label='KNN sul Train set')
plt.plot(k_neighbors, acc_test_raw,  label='KNN sul Test set')
plt.plot(k_neighbors, acc_test_pca,  label='KNN sul Test set (Con PCA)')
plt.plot(k_neighbors, acc_test_mda,  label='KNN sul Test set (Con MDA)')

plt.title('Analisi dell\'Accuracy del KNN al variare di K (voti pesati sulla distanza)')
plt.xlabel('Numero di vicini (k)')
plt.ylabel('Accuracy')
plt.xticks(np.arange(0, 51, 5))
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='upper right')

plt.show()
```

```python
#Ho fatto delle prove con i valori di k compresi tra 5 e 10: 7 è risultato il migliore e quindi l'ho scelto

# Predizione sui dati raw
knn_raw = KNN(n_neighbors=7, weights='distance')
knn_raw.fit(X_train_scaled, y_train)
y_pred_knn_train = knn_raw.predict(X_train_scaled)
y_pred_knn = knn_raw.predict(X_test_scaled)
    
#Predizione sui dati con PCA
knn_pca = KNN(n_neighbors=7, weights='distance')
knn_pca.fit(X_pca_train, y_train) 
y_pred_knn_PCA = knn_pca.predict(X_pca_test)
    
#Predizioni sui dati con MDA
knn_mda = KNN(n_neighbors=7, weights='distance')
knn_mda.fit(Zmda_train, y_train)
y_pred_knn_MDA = knn_mda.predict(Zmda_test)

accuracy_dict = {"KNN sul Train": [accuracy_score(y_train, y_pred_knn_train), precision_score(y_train, y_pred_knn_train, average='weighted'), recall_score(y_train, y_pred_knn_train, average='weighted'), f1_score(y_train, y_pred_knn_train, average='weighted')],
                 "KNN sul Test": [accuracy_score(y_test, y_pred_knn), precision_score(y_test, y_pred_knn, average='weighted'), recall_score(y_test, y_pred_knn, average='weighted'), f1_score(y_test, y_pred_knn, average='weighted')],
            "KNN sul Test (PCA)": [accuracy_score(y_test, y_pred_knn_PCA), precision_score(y_test, y_pred_knn_PCA, average='weighted'), recall_score(y_test, y_pred_knn_PCA, average='weighted'), f1_score(y_test, y_pred_knn_PCA, average='weighted')],
                 "KNN sul Test (MDA)": [accuracy_score(y_test, y_pred_knn_MDA), precision_score(y_test, y_pred_knn_MDA, average='weighted'), recall_score(y_test, y_pred_knn_MDA, average='weighted'), f1_score(y_test, y_pred_knn_MDA, average='weighted')]
           }
accuracy = pd.DataFrame(accuracy_dict, index=['Accuracy', 'Precision', 'Recall', 'F1-score'])
display(accuracy)
```
